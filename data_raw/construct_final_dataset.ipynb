```python
import pandas as pd
import requests
```

```python
def neighborhood_from_coordinates(lon, lat):
    mapbox_key = "pk.eyJ1IjoibXBlcmV6MDIwNyIsImEiOiJjanZlZTJnOTMwbmltNDRwbDBjajdsNG84In0.kd-GVttMUZcdM3rFc6GJjg"
    mapbox_url = "https://api.mapbox.com/geocoding/v5/mapbox.places/" + str(lon) + "," + str(lat) + ".json";
    mapbox_params = {"access_token":mapbox_key, "types":"neighborhood"}
    
    response = requests.get(mapbox_url, params=mapbox_params).json()
    
    try:
        neighborhood = response["features"][0]["text"].lower().title()
        borough = response["features"][0]["context"][0]["text"].lower().title()
        
    except IndexError:
        features = response["features"]
        
        if len(features) != 0:
            print(features)
            
        neighborhood = ""
        borough = ""
    
    return neighborhood, borough
```

```python
laundry_df = pd.read_csv("nyc_laundry.csv")
laundry_df.head()
```

```python
laundry_df = laundry_df.drop_duplicates(["Longitude", "Latitude"])
laundry_df.head()
```

```python
unpackdf=pd.DataFrame(laundry_df.apply(lambda x : neighborhood_from_coordinates(x["Longitude"],x["Latitude"]),axis=1).tolist(),
                      columns=["Neighborhood", "Borough"],index=laundry_df.index)
laundry_df = pd.concat([laundry_df, unpackdf],axis=1)

laundry_df.head()
```

```python
laundry_df = laundry_df[laundry_df["Borough"] != ""] # drop rows with empty neighborhood/borough

laundry_counts = laundry_df["Neighborhood"].value_counts()
laundry_counts.name = "Laundromats"
neighborhoods_df = pd.DataFrame(laundry_counts)
neighborhoods_df.index.name = "Neighborhood"
neighborhoods_df.head()
```

```python
def define_rent_categories(rent):
    rent_category = "Low Rent"
    if rent > 6000:
        rent_category = "High Rent"
    elif rent > 4000:
        rent_category = "Medium Rent"
        
    return rent_category
```
